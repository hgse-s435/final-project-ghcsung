```python
# using glob, find all the data files
import glob

files = glob.glob('./*.csv')
print(files)
```

```python
# read in as DataFrame
import pandas as pd
list_df = []

for filename in files: 
    list_df.append(pd.read_csv(filename, sep=',', error_bad_lines=False, index_col=False, dtype='unicode'))
```

```python
list_df[0]
```

```python
# Drop columns: Temp: unreliable, tags: not used, ACC: ???, IBI & BVP: proxy of HR
list_df2 = []
droplist = ["TEMP","TEMP real time", "tags", "tags real time", "HR real time", 
           "ACC real time", "ACC_x", "ACC_y","ACC_z", "IBI", "IBI real time","BVP","BVP real time"]
for df in list_df:
    list_df2.append(df.drop(droplist, axis = 1))
```

```python
list_df2[0]
```

```python
# pad the HR using ffill (forward fill), then drop all columns with NaN

list_df3 = []
for df in list_df2:
    df.HR = df.HR.fillna(method = 'ffill')
    list_df3.append(df.dropna())
                    
```

```python
list_df3[0]
```

```python
list_df3[0].dtypes
```

```python
# use EDA real time as index
list_df4 = []
for df in list_df3:
    realtime = pd.to_datetime(df["EDA real time"])
    df.set_index(realtime, inplace = True)
    list_df4.append(df)
```

```python
# final clean
for df in list_df4:
    df = df.dropna()
    df = df.drop("EDA real time", axis = 1)
```
